# P2 Data Processing - Rentals

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from glob import glob

pd.set_option("display.max_columns", 100)
```

```python
DATA_DIR = os.path.join(os.environ["HOME"], "Datasets", "HTX_Hackathon","ps2-rental_bases")
```

# Read Data

```python
p2_listing = pd.read_csv(os.path.join(DATA_DIR, "listings.csv"))
p2_reviews = pd.read_csv(os.path.join(DATA_DIR, "reviews.csv"))

display(p2_listing.shape)
display(p2_listing.head())
display(p2_reviews.shape)
display(p2_reviews.head())
```

## Clean Up `listing_id` Column in Reviews

```python
p2_reviews_clean = p2_reviews.copy()
p2_reviews_clean = p2_reviews_clean[p2_reviews_clean["listing_id"].apply(lambda x: x.isnumeric())]

display(p2_reviews_clean.shape)
display(p2_reviews_clean.head())
```

# Join Data

```python
# convert join keys to int
p2_reviews_clean["listing_id"] = p2_reviews_clean["listing_id"].astype(int)
p2_listing["id"] = p2_listing["id"].astype(int)

p2_joined = p2_listing.merge(p2_reviews_clean, how = "left",
                            left_on = "id", right_on = "listing_id")

display(p2_joined.shape)
display(p2_joined.head())
```

```python
# CSV baseline data
p2_joined.to_csv(os.path.join(DATA_DIR, "P2_Listing_Base.csv"), index = False)
```
